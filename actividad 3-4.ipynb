**Abstracto**

La calidad del aire urbano es un determinante crucial de la salud pública, especialmente en las extensas ciudades de los Estados Unidos. Este estudio busca analizar la relación entre las condiciones de calidad del aire y las variables meteorológicas con los riesgos para la salud pública en diversas ciudades estadounidenses utilizando el dataset "Urban Air Quality and Health Impact Dataset". Mediante un análisis exploratorio de datos (EDA) y modelos predictivos, se identifican patrones y correlaciones significativas que afectan el puntaje de riesgo para la salud (Health_Risk_Score). Los hallazgos revelan que factores como el índice de calor, la humedad y las ráfagas de viento tienen una influencia notable en los riesgos para la salud. Este análisis está dirigido a formuladores de políticas públicas, autoridades sanitarias y entidades privadas interesadas en mejorar la calidad de vida urbana mediante estrategias basadas en evidencia. Las recomendaciones derivadas buscan mitigar los impactos negativos de la contaminación del aire, promoviendo entornos urbanos más saludables y sostenibles.

# Figura 2: Boxplot de Temperatura Promedio Diaria (temp)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Season", y="temp", data=df, palette="Set2")
plt.title("Boxplot de Temperatura Promedio Diaria por Temporada")
plt.xlabel("Temporada")
plt.ylabel("Temperatura Promedio (°F)")
plt.show()
```

    palette: Define la paleta de colores para mejorar la visualización.

   **Distribución de la Temperatura Promedio en Otoño (Fall)**:
- El boxplot muestra la temperatura promedio diaria durante la temporada de otoño, con una mediana alrededor de los 75 °F.
- El rango intercuartílico (Q1 a Q3) oscila entre 70 °F y 85 °F, lo que indica que la mayor parte de los días tienen temperaturas dentro de este rango.
- Los valores máximos y mínimos son 60 °F y 100 °F, lo que sugiere días con condiciones climáticas más extremas dentro de la temporada de otoño, aunque los valores cercanos a estos extremos son menos frecuentes.

# Figura 3: Heatmap de Correlación entre Variables Meteorológicas y Health_Risk_Score

Este heatmap muestra las correlaciones entre las variables numéricas y Health_Risk_Score, identificando las más influyentes.

```python
# Selecciono solo columnas numéricas relevantes
numericas = df.select_dtypes(include=['float64', 'int64'])

# Calculo la matriz de correlación
corr = numericas.corr()

# Selecciono las correlaciones con 'Health_Risk_Score'
corr_health = corr[['Health_Risk_Score']].sort_values(by='Health_Risk_Score', ascending=False)

# Creo el heatmap
plt.figure(figsize=(10, 12))
sns.heatmap(corr_health, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlación entre Variables y Health_Risk_Score")
plt.show()
```

**Descripción**:

    select_dtypes: Filtra solo las columnas numéricas para calcular la correlación.
    sns.heatmap: Crea un mapa de calor de las correlaciones, destacando las más fuertes.

Esta figura pretende identificar qué variables metereológicas tienen una mayor correlación con el Helath risk score, los valores cercanos a 1 o -1 indican una fuerte correlación positiva o negativa, respectivamente.
Segun la grafica se puede identificar que el índice de calor, la puntuación de severidad, ráfagas de viento y el rocío tienen una relación positiva y que variables como la visivilidad, dirección del viento y la presión tienen una relacion negativa.

# Figura 4: Relación entre Humidity (humidity) y Health_Risk_Score

Este scatter plot explora la posible relación entre la humedad relativa y el riesgo para la salud pública.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="humidity", y="Health_Risk_Score", hue="Season", data=df, palette="viridis", alpha=0.6)
plt.title("Relación entre Humedad Relativa y Health_Risk_Score")
plt.xlabel("Humedad Relativa (%)")
plt.ylabel("Health Risk Score")
plt.legend(title='Temporada')
plt.show()
```

**Descripción**:

    sns.scatterplot: Crea un diagrama de dispersión con humidity en el eje X y Health_Risk_Score en el eje Y.
    hue: Diferencia los puntos por temporada (Season) para ver si hay patrones estacionales.
    alpha: Define la transparencia de los puntos para manejar la superposición.

Esta figura explora si existe una relación entre la humedad relativa y el riesgo para la salud. Una tendencia ascendente en el scatter plot sugeriría que mayores niveles de humedad están asociados con mayores riesgos de salud. Segun se identifica en la tendecia ascendente puede sugerir que a mayores niveles de humedad están asociados a mayores riesgos de la salud.

# Figura 5: Impacto Combinado de Wind_Speed (windspeed) y Temp_Avg (temp) en Health_Risk_Score

Este scatter plot analiza cómo la velocidad del viento y la temperatura promedio interactúan para influir en el riesgo para la salud.

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x="windspeed", y="temp", hue="Health_Risk_Score", size="Health_Risk_Score",
                          sizes=(20, 200), palette="coolwarm", data=df, alpha=0.7)
plt.title("Impacto de Velocidad del Viento y Temperatura en Health_Risk_Score")
plt.xlabel("Velocidad del Viento (mph)")
plt.ylabel("Temperatura Promedio (°F)")
plt.legend(title='Health Risk Score', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Descripción**:

    sns.scatterplot: Diagrama de dispersión con windspeed en el eje X y temp en el eje Y.
    hue y size: Ambos mapeados a Health_Risk_Score para representar el riesgo mediante color y tamaño de los puntos.
    sizes: Define el rango de tamaños para los puntos.
    alpha: Transparencia para manejar la superposición de puntos.

Esta figura analiza cómo la velocidad del viento y la temperatura promedio interactua para influir en el riesgo para la salud.
Los patrones podrian indicar que cuando se combina altas velocidades del viento y las altas temperaturas existe un aumento del riesgo de la salud.
